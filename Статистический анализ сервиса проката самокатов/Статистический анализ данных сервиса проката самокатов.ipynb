# Статистический анализ данных аренды самокатов "GoFast"

## Цель исследования



Основная задача исследования: 
>Проверить ряд гипотез, которые могут помочь увеличить количество пользователей с подпиской.

Дополнительно:
 - определить необходимое количество промокодов для проведения акции для увеличения количества пользователей с подпиской, плановый показатель - $10\%$ пользователей продлили подписку.
 - оценить охват push-уведомлений, плановый показатель - $399.5$ тыс. пользователей.
 


Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. 

Сервисом можно пользоваться:
 - без подписки
    - абонентская плата отсутствует;
    - стоимость одной минуты поездки — $8$ рублей;
    - стоимость старта (начала поездки) — $50$ рублей;
    
    
 - с подпиской Ultra
    - абонентская плата — $199$ рублей в месяц;
    - стоимость одной минуты поездки — $6$ рублей;
    - стоимость старта — бесплатно.

## План исследований

### Описание данных

В основных данных есть информация о пользователях, их поездках и подписках.

**Пользователи** — `users_go.csv`

 - `user_id` - уникальный идентификатор пользователя
 - `name` - имя пользователя
 - `age` - возраст
 - `city` - город
 - `subscription_type` - тип подписки (free, ultra)
 
**Поездки** — `rides_go.csv`

 - `user_id` - уникальный идентификатор пользователя
 - `distance` - расстояние, которое пользователь проехал в текущей сессии (в метрах)
 - `duration` - продолжительность сессии (в минутах) — время с того момента, как пользователь нажал кнопку «Начать поездку» до момента, как он нажал кнопку «Завершить поездку»
 - `date` - дата совершения поездки
 
**Подписки** — `subscriptions_go.csv`
 - `subscription_type` - тип подписки
 - `minute_price` - стоимость одной минуты поездки по данной подписке
 - `start_ride_price` - стоимость начала поездки
 - `subscription_fee` - стоимость ежемесячного платежа

### Шаг 6. Проверка гипотез

>Продакт-менеджеры сервиса хотят увеличить количество пользователей с подпиской. Для этого они будут проводить различные акции, но сначала нужно выяснить несколько важных моментов.
 - 6.1 Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверьте гипотезу. Используйте исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.
 
     **Метод:** Двухвыборочный t-тест для независимых выборок (`ttest_ind()`)
 > - $H_0: \mu_1 = \mu_2$
 > - $H_1: \mu_1 > \mu_2$, где
 >> - $\mu_1$ - среднее время поездок пользователей с подпиской
 >> - $\mu_2$ - среднее время поездок пользователей без подписки
 
 
 - 6.2 Расстояние одной поездки в $3130$ метров — оптимальное с точки зрения износа самоката. Можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает $3130$ метров? Проверьте гипотезу и сделайте выводы.
 
     **Метод:** Одновыборочный t-тест о равенстве среднего ГС числу
 > - $H_0: \mu_1 = 3130$

Для оценки данных параметров будем использовать следующий алгоритм:

 2. далее расмотрим распредение на графике.
 
Для этого используем следующую функцию для описания данных:
 - размах значений:

```python
def range_values(series:pd.Series) -> str:
    try:
        i_QR = st.iqr(series)
        low_span_lim = series.quantile(0.25) - 1.5 * i_QR
        high_span_lim = series.quantile(0.75) + 1.5 * i_QR
        if low_span_lim < series.describe()['min']:
            low_span_lim = series.describe()['min']
        if high_span_lim > series.describe()['max']:
            high_span_lim = series.describe()['max']
        return f'{int(low_span_lim)} <= {series.name} <= {int(high_span_lim)}'
    except ValueError as val_ex:
        print('Возникло исключение:', str(val_ex))
        return ''
        
```

## Проверка гипотез

Продакт-менеджеры сервиса хотят увеличить количество пользователей с подпиской. Для этого они будут проводить различные акции, но сначала нужно выяснить несколько важных моментов.

###  Время использования приложения пользователями разных категорий

 Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверьте гипотезу. Используйте исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.
 
   **Метод:** Двухвыборочный t-тест для независимых выборок (`ttest_ind()`)
 > - $H_0: \mu_1 = \mu_2$
 > - $H_1: \mu_1 \ge \mu_2$, где
 >> - $\mu_1$ - среднее время поездок пользователей с подпиской
 >> - $\mu_2$ - среднее время поездок пользователей без подписки

Для проверки данной гипотезы будем использовать ранее обозначенные выборки:
 - `ultra_duration`: содержащие данные о длительности поездок пользователей с подпиской ULTRA
 - `free_duration`: содержащие данные о длительности поездок пользователей без подписки
 
Изначально мы знаем, что размеры выборок отличаются, поэтому проверим значения дисперсии выборок, для определения параметра `equal_var`

```python
print('Дисперсия выборки пользователей с подпиской: ', ultra_duration.var())
```

```python
print('Дисперсия выборки пользователей без подписки: ',free_duration.var())
```

Т.к. дисперсии не равны, установим параметр `equal_var = False`

```python
alpha = 0.05 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

results = st.ttest_ind(ultra_duration, free_duration, alternative='greater', equal_var=False)

print('p-значение', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нулевая гипотеза не отвергнута')
```

Отвергнув нулевую гипотезу мы обозначаем, что подтверждаем, что среднее значение длительности поездок пользователей с подпиской ULTRA больше, чем у пользователей без подписки. Частично данный тезис мы выдвигали в пункте [5.2 - Распределение длительности поездки по категориям подписки](#Распределение_длительности)

 > - $H_0: \mu_1 = 3130$
 > - $H_1: \mu_1 \le{3130}$, где
 >> - $\mu_1$ - среднее расстояние поездки пользователей с подпиской
 
 
Для проверки данной гипотезы используем выборку `ultra_distance`

### Акция "Бесплатный месяц"

Отделу маркетинга GoFast поставили задачу:

>нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум  100 существующих клиентов должны продлить эту подписку.

То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают  10%
  пользователей.

Выясните, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно  5%
 . Подберите параметры распределения, описывающего эту ситуацию, постройте график распределения и сформулируйте ответ на вопрос о количестве промокодов.

В данном случае исходами эксперимента можно считать следующее:
 - пользователь продлил подписку — успех;
 - пользователь не стал продливать подписку — неудача.
 
Нам известна вероятность успеха в предыдущих наблюдениях: $p=10\%$

Следовательно вероятность неудачи: $1 - p = 90\%$

Необходимо разослать $n$ писем, чтобы по итогу как минимум 100 пользователей продлило подписку с вероятностью $p\approx{95\%}$



```python
p = 0.1 # вероятность успеха
max_val = 1000
k = 100
n_val_dict = {}
distr = []
low_lim = 0.045
hight_lim = 0.055
target_val = 0.05

print(f'Список значений количества рассылаемых предложений и вероятность не достижения значения 100 новых подписчиков:\n') 
      
for i in range(100, max_val):
    binom_distr = st.binom(n=i, p=p)
    result = 1-binom_distr.cdf(k)
    
    if low_lim < result < hight_lim:
        print(i, result)
        n_val_dict[target_val - result] = i

err_list = n_val_dict.keys()

closest = min(err_list, key=lambda x: abs(x - 0))
N = n_val_dict.get(closest)
print(
    f'\nОтпимальное количество промокодов минимально необходимое для выполнения плана с вероятностью в 95% : {N}'
)   
    
```

На основе данного значения построим визуализацию:

```python
plt.figure(figsize=(15, 6))

distr = []

for k in range(0, N + 1):
    current_value = binom.pmf(k, N, p)
    distr.append(current_value)
    
plt.bar(range(0, N + 1), distr)
plt.title('Распределение вероятности получения 100 новых подписчиков при рассылке 858 промокодов', fontsize = 15)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.xlabel('Количество промокодов', fontsize=12)
plt.show()
```

**Отпимальное количество промокодов минимально необходимое для выполнения плана с вероятностью в $95\%$ : $858$**

### Оценка эффективности push-уведомлений

Отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении. Клиенты могут открыть его или не открывать. Известно, что уведомления открывают около $40\%$ получивших клиентов. Отдел планирует разослать $1$ млн уведомлений.

С помощью аппроксимации постройте примерный график распределения и оцените вероятность того, что уведомление откроют не более $399,5$ тыс. пользователей.

В данном случае:
 - $N = 1 000 000$
 - $p = 0.4$
 - $k = 399 500$
 
Т.к. значение вероятности $p = 40\%$, следовательно необходимо применить апроксимацию биноминального распределения к нормамальному.
Для этого рассчитаем мат. ожидание $\mu$ и стандартное отклонение $\sigma$, т.к. нормальное распределение задается двумя этими параметрами:
 > $N(\mu, \sigma^2)$

```python
n = 1000000 
p = 0.4 
target_val = 399500

mu = n*p 
sigma = sqrt(n*p*(1-p)) 

distr = st.norm(mu, sigma) 

result = 1- distr.cdf(target_val)  


left = int(mu - 3 * sigma)
right = int(mu + 3 * sigma)

def linspace(start, stop, num):
    step = (stop - start) / (num - 1)
    result = []
    for i in range(num):
        result.append(start + step * i)
    return result


plt.figure(figsize=(15,6))
ax = plt.subplot()
x_1 = list(range(max(0, left), right))
fig_1 = ax.plot(x_1, binom.pmf(x_1, n, p), 'bo', ms=8, label='вероятности биномиального распределения')

x_2 = linspace(left, right, 100)
fig_2 = ax.plot(x_2, norm.pdf(x_2, mu, sigma), 'r-', lw=3, alpha=1, label='плотность вероятности нормального распределения')

y_2 = linspace(0, 0.0005, 5)
x_ticks = list(range(int(mu - 3 * sigma), int(mu + 3 * sigma), 500))
ax.set_xticks(x_ticks)

x_p = linspace(target_val, right, 10)
ax.fill_between(x_p, norm.pdf(x_p, mu, sigma), 0, label='Плотность вероятности $P(X=399500)$')
plt.legend()
plt.show()


if (mu - (3 * sigma) >= 0) and ((mu + (3 * sigma) <= n)):
    print('Условие выполняется')
else:
    print('Условие не выполняется')
print(f'Плотность вероятности точки {target_val} = {result}')
```
